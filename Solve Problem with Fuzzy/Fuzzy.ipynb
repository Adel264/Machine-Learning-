```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

# ***Inputs variables***

```python
x_Cholestrol =ctrl.Antecedent( np.arange(0,700,1),'cholestrol')
x_BloodPressure =ctrl.Antecedent( np.arange(0,250,1), 'Blood Pressure')
x_Age =ctrl.Antecedent( np.arange(0,80,1) , 'Age')
x_BMI =ctrl.Antecedent( np.arange(0,70,1) , 'BMI')
x_Diabetes =ctrl.Antecedent( np.arange(0,550,1) , 'Diabetes' )
x_HeartDisease =ctrl.Consequent( np.arange(0,6,1) , 'Heart Disease')
```

```python
x_Cholestrol['normal'] = fuzz.trapmf(x_Cholestrol.universe, [0, 0, 50, 200])
x_Cholestrol['medium'] = fuzz.trimf(x_Cholestrol.universe, [190, 220, 250])
x_Cholestrol['high'] = fuzz.trimf(x_Cholestrol.universe, [220, 260, 310])
x_Cholestrol['very high']= fuzz.trapmf(x_Cholestrol.universe, [280, 360, 700, 700])
```

```python
x_Cholestrol.view()
```

```python
x_BloodPressure['normal'] = fuzz.trapmf(x_BloodPressure.universe, [0, 0, 67, 130])
x_BloodPressure['medium'] = fuzz.trimf(x_BloodPressure.universe, [120, 130, 159])
x_BloodPressure['very high']= fuzz.trapmf(x_BloodPressure.universe, [149, 165, 250, 250])
```

```python
x_BloodPressure.view()
```

```python
x_Age['young'] = fuzz.trapmf(x_Age.universe, [0, 0, 21, 38])
x_Age['middle age'] = fuzz.trimf(x_Age.universe, [33, 40,45])
x_Age['old'] = fuzz.trimf(x_Age.universe, [40, 48, 59])
x_Age['very old'] = fuzz.trapmf(x_Age.universe, [52, 60, 80, 80])
```

```python
x_Age.view()
```

```python
x_BMI['normal'] = fuzz.trapmf(x_BMI.universe, [0, 0, 10, 26])
x_BMI['over weight'] = fuzz.trimf(x_BMI.universe, [25, 26.5, 33])
x_BMI['obese']= fuzz.trapmf(x_BMI.universe, [30, 40, 70, 70])
```

```python
x_BMI.view()
```

```python
x_Diabetes['normal'] = fuzz.trimf(x_Diabetes.universe, [0, 70, 160])
x_Diabetes['diabetic'] = fuzz.trimf(x_Diabetes.universe, [140, 230, 400])
```

```python
x_Diabetes.view()
```

# ***Output variable***

```python
x_HeartDisease['healthy'] = fuzz.trapmf(x_HeartDisease.universe, [0, 0, 0.4, 1.7])
x_HeartDisease['early-stage'] = fuzz.trimf(x_HeartDisease.universe, [1.5, 2, 2.5])
x_HeartDisease['advanced-stage']= fuzz.trapmf(x_HeartDisease.universe, [2.4, 3, 5, 5])
```

```python
x_HeartDisease.view()
```

# ***Rule Base***

```python
rule1 = ctrl.Rule(x_Age['young'] & x_BloodPressure['normal'] & x_Cholestrol['normal'] & x_Diabetes['normal'] | x_BMI['obese'] ,x_HeartDisease['healthy']  )
rule2 = ctrl.Rule(x_Age['very old'] & x_BloodPressure['normal'] & x_Cholestrol['normal'] & x_Diabetes['normal'] | x_BMI['normal'] ,x_HeartDisease['early-stage']  )
rule3 = ctrl.Rule(x_Age['old'] & x_BloodPressure['very high'] | x_Cholestrol['normal'] & x_Diabetes['diabetic'] & x_BMI['normal']  ,x_HeartDisease['early-stage']  )
rule4 = ctrl.Rule(x_Age['young'] & x_BloodPressure['medium'] | x_Cholestrol['normal'] & x_Diabetes['diabetic'] & x_BMI['over weight'] ,x_HeartDisease['early-stage']  )
rule5 = ctrl.Rule(x_Age['middle age'] & x_BloodPressure['normal'] & x_Cholestrol['high'] & x_Diabetes['diabetic'] & x_BMI['over weight']  ,x_HeartDisease['advanced-stage']  )
rule6 = ctrl.Rule(x_Age['old'] & x_BloodPressure['very high'] | x_Cholestrol['high'] & x_Diabetes['normal'] | x_BMI['obese']  ,x_HeartDisease['advanced-stage']  )
rule7 = ctrl.Rule(x_Age['middle age'] & x_BloodPressure['medium'] & x_Cholestrol['very high'] & x_Diabetes['diabetic'] & x_BMI['over weight'] ,x_HeartDisease['advanced-stage']  )
rule8 = ctrl.Rule(x_Age['old'] & x_BloodPressure['medium'] & x_Cholestrol['very high'] & x_Diabetes['diabetic'] & x_BMI['obese'],x_HeartDisease['advanced-stage']  )
rule9 = ctrl.Rule(x_Age['very old'] & x_BloodPressure['very high'] & x_Cholestrol['very high'] & x_Diabetes['diabetic'] & x_BMI['over weight']  ,x_HeartDisease['advanced-stage']  )
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

```python
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```

```python
tipping.input['cholestrol'] = 300
tipping.input['Blood Pressure'] = 130 
tipping.input['Age'] = 40
tipping.input['BMI'] = 32
tipping.input['Diabetes'] = 230

tipping.compute()
```

```python
print (tipping.output['Heart Disease'])
x_HeartDisease.view(sim=tipping)
```
